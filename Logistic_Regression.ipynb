```python
import pandas as pd

data = data.copy()

Q1 = data['Price'].quantile(0.25)
Q3 = data['Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR 

data_no_outliers = data[(data['Price'] >= lower_bound) & (data['Price'] <= upper_bound)]

print("Original dataset size:", data.shape)
print("Dataset size after removeing outliers:", data_no_outliers.shape)

import seaborn as sns 
import matplotlib.pyplot as plt

sns.boxplot(x=data_no_outliers['Price'])
plt.title("Price Boxplot with no Outliers")
plt.show()
```

```python
# Removing Mileage Outliers using IQR
Q1 = data["Mileage"].quantile(0.25)
Q3 = data["Mileage"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

data = data[(data["Mileage"] >= lower) & (data["Mileage"] <= upper)]

print("Shape after removing Mileage outliers:", data.shape)
```

```python
data.head()
```

```python
# Removing Price Outliers using IQR
Q1 = data["Engine volume"].quantile(0.25)          
Q3 = data["Engine volume"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

data = data[(data["Engine volume"] >= lower) & (data["Engine volume"] <= upper)]

print("Shape after removing Engine volume outliers:", data.shape)
```

```python
data.head()
```

```python
data["Doors"] = data["Doors"].replace({
    "04-May": 4,
    "02-Mar": 2,
    ">5": 5
})

data["Doors"] = pd.to_numeric(data["Doors"])
```
